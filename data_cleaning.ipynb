```python
neutral_home_winner_indices, neutral_home_loser_indices = np.zeros(len(winner_away), dtype=bool), np.zeros(len(winner_away), dtype=bool)
```

```python
import random
```

```python
for idx, truth_val in enumerate(neutral_indices):
    if truth_val:
        # If it's a neutral game, do a 50/50 coin toss whether to force the home team to be the winner
        # team or the loser team.
        if random.randint(0,1) == 1:
            neutral_home_winner_indices[idx] = True
        else:
            neutral_home_loser_indices[idx] = True
```

```python
np.all(neutral_home_loser_indices ^ neutral_home_winner_indices == neutral_indices)
```

```python
np.all(neutral_home_loser_indices ^ neutral_home_winner_indices ^ winner_home ^ winner_away)
```

```python
def swap(df, stat):
    df.loc[winner_home, f'H{stat}'] = df.loc[winner_home, f'W{stat}']
    df.loc[loser_home, f'H{stat}'] = df.loc[loser_home, f'L{stat}']
    df.loc[winner_away, f'A{stat}'] = df.loc[winner_away, f'W{stat}']
    df.loc[loser_away, f'A{stat}'] = df.loc[loser_away, f'L{stat}']
    df.loc[neutral_home_winner_indices, f'H{stat}'] = df.loc[neutral_home_winner_indices, f'W{stat}']
    df.loc[neutral_home_winner_indices, f'A{stat}'] = df.loc[neutral_home_winner_indices, f'L{stat}']
    df.loc[neutral_home_loser_indices, f'H{stat}'] = df.loc[neutral_home_loser_indices, f'L{stat}']
    df.loc[neutral_home_loser_indices, f'A{stat}'] = df.loc[neutral_home_loser_indices, f'W{stat}']
    df.drop([f'W{stat}', f'L{stat}'], axis=1, inplace=True)

for stat in stats + ['TeamID', 'Score']:
    swap(reg_season, stat)

reg_season.head()
```

```python
positive_self_offense_attrs = ['FGM', 'FGA', 'FTA', 'FTM', 'FGM3', 'FGA3','OR', 'Ast', 'Score']
```

```python
positive_self_defense_attrs = ['Stl', 'Blk', 'PF', 'DR']
positive_oppt_defense_attrs = ['TO', 'FTA', 'OR', 'Score']
```

```python
from tensordict import TensorDict
```

```python
def home_stats(stats):
    return list(map(lambda x : f'H{x}', stats))

def away_stats(stats):
    return list(map(lambda x : f'A{x}', stats))
```

```python
pd.options.mode.copy_on_write = True
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence
```

```python
def get_team_ids_in_order(df):
    return list(sorted(set(df['HTeamID'].unique()).union(set(df['ATeamID']))))
```

```python
number_of_teams = len((get_team_ids_in_order(scores)))
```

```python
def get_sequence_data(df):
    seasons = df['Season'].unique()
    defense_game_data_per_season = {}
    offense_game_data_per_season = {}
    for season in seasons:
        games_in_season = df[df['Season'] == season]
        offense_games_in_current_season = [torch.zeros(1)] * number_of_teams
        defense_games_in_current_season = [torch.zeros(1)] * number_of_teams
        for team in get_team_ids_in_order(games_in_season):
            home_games = games_in_season[games_in_season['HTeamID'] == team]
            offense_home_stats_cols = home_stats(positive_self_offense_attrs)
            offense_home_stats = home_games[['DayNum'] + offense_home_stats_cols]

            offense_home_stats.rename(columns={col : new_col for col, new_col in zip(offense_home_stats_cols, positive_self_offense_attrs)}, inplace=True)

            defense_home_stats_cols = home_stats(positive_self_defense_attrs) + away_stats(positive_oppt_defense_attrs)
            defense_home_stats = home_games[['DayNum'] + defense_home_stats_cols]
            defense_cols = positive_self_defense_attrs + positive_oppt_defense_attrs
            defense_home_stats.rename(columns={col : new_col for col, new_col in zip(defense_home_stats_cols, defense_cols)}, inplace=True)

            away_games = games_in_season[games_in_season['ATeamID'] == team]
            offense_away_stats_cols = away_stats(positive_self_offense_attrs)
            offense_away_stats = away_games[['DayNum'] + offense_away_stats_cols]

            offense_away_stats.rename(columns={col : new_col for col, new_col in zip(offense_away_stats_cols, positive_self_offense_attrs)}, inplace=True)
            defense_away_stats_cols = away_stats(positive_self_defense_attrs) + home_stats(positive_oppt_defense_attrs)
            defense_away_stats = away_games[['DayNum'] + defense_away_stats_cols]
            defense_away_stats.rename(columns={col : new_col for col, new_col in zip(defense_away_stats_cols, defense_cols)}, inplace=True)
            offense_stats = pd.concat([offense_home_stats, offense_away_stats])
            offense_stats.sort_index(level='DayNum', inplace=True, ascending=True)

            offense_stats.drop(columns=['DayNum'], inplace=True)

            defense_stats = pd.concat([defense_home_stats, defense_away_stats])
            defense_stats.sort_index(level='DayNum', inplace=True, ascending=True)

            defense_stats.drop(columns=['DayNum'], inplace=True)

            assert len(offense_stats) == len(defense_stats)        
            team = int(team)
            offense_games_in_current_season[team] = torch.Tensor(offense_stats.to_numpy()).float()
            defense_games_in_current_season[team] = torch.Tensor(defense_stats.to_numpy()).float()
    
        offense_game_data_per_season[str(season)] = pad_sequence(offense_games_in_current_season, batch_first=True)
        defense_game_data_per_season[str(season)] = pad_sequence(defense_games_in_current_season, batch_first=True)
    return offense_game_data_per_season, defense_game_data_per_season
```

```python
train_offense_sequence_data_per_season, train_defense_sequence_data_per_season = get_sequence_data(train_data)
```

```python
offense_data = TensorDict(train_offense_sequence_data_per_season, batch_size=number_of_teams)
torch.save(offense_data, "./train_offense_sequence_data.pt")
```
